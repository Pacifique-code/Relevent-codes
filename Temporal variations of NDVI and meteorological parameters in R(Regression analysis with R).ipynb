# Plotting relation between start, length, end of rainy season and NDVI based on CHIRPS dataset

```r
#Data = read.csv('START_RAIN_NEB2.csv') # Reading data
Data = read.csv('Yearly_NEB_data1.csv')
# Reading dat from github
#library(curl)
#data = read.csv(curl('https://github.com/Pacifique-code/Thesis_Data/blob/master/Data/START_RAIN_NEB2.csv'))
```

```r
head(Data)
```

```r
library("basicTrendline") #c("lin2P","line3P","log2P","exp2P","exp3P","power2P","power3P")

par(mfrow=c(1,1)) 
trendline(Data$CHIRPS.start, Data$NDVI, model = "line2P", linecolor = "red", lty = 1, lwd = 2,CI.fill = FALSE, 
          CI.color = "black",CI.lty = 2,summary = F, ylab = 'NDVI', xlab= 'Onset date (days from first July)')

trendline(Data$	CHIRPS.end, Data$NDVI, model = "line2P", linecolor = "red", lty = 1, lwd = 2,
          CI.fill = FALSE, CI.color = "black",CI.lty = 2,summary = F,ylab = 'NDVI', xlab = 'End of growing season (days from first July)')

trendline(Data$CHIRPS.length, Data$NDVI, model = "line2P",  linecolor = "red", lty = 1, lwd = 2,
          CI.fill = FALSE, CI.color = "black",CI.lty = 2,ylab = 'NDVI',summary = F, xlab= 'Length of growing season (number of days)')

dev.copy(jpeg,'seasons_NDVI_regression.jpeg')
dev.off()
```

# Plotting SOND, MAM, Annual rainfall and NDVI-mean

Here we pair seasonal rainfall total and mean-ndvi

```r
#Seasonal_data = read.csv('NEB_SEASON_PREC2.csv',header =TRUE)
#head(Seasonal_data)
```

```r
trendline(Data$SOND_CHIRPS, Data$SOND_NDVI, model = "line2P", linecolor = "red", lty = 1, lwd = 2,CI.fill = FALSE, 
          CI.color = "black",CI.lty = 2,summary = F, ylab = 'SOND NDVI', xlab= 'MAM Precipitation (mm))')

trendline(Data$MAM_CHIRPS, Data$MAM_NDVI, model = "line2P", linecolor = "red", lty = 1, lwd = 2,
          CI.fill = FALSE, CI.color = "black",CI.lty = 2,summary = F,ylab = 'MAN NDVI', xlab = 'MAM Precipitation (mm)')

trendline(Data$CHIRPS_prec, Data$NDVI, model = "line2P",  linecolor = "red", lty = 1, lwd = 2,
          CI.fill = FALSE, CI.color = "black",CI.lty = 2,ylab = 'Annual NDVI mean',summary = F, xlab= 'Annual Precipitation (mm)')

dev.copy(jpeg,'seasonal_NDVI_regression.jpeg')
dev.off()
```
