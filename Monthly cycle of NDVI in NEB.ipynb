Comparing precipitation with NDVI for Kirundo

```python
#create list of months
Month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 
         'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#create list for made up average percipitation for kirundo observed data %
Kirundo_obs_pre = [81.941 ,98.725 ,136.191 ,184.609 ,100.122 ,16.547 ,7.122 ,22.688 ,70.772 ,114.178 ,127.041 ,92.516 
]
#create list for made up average percipitation for muyinga observed data %
Muyinga_obs_prec = [110.113 ,110.003 ,157.448 ,180.673 ,74.244 ,14.800 ,1.572 ,17.003 ,54.431 ,106.178 ,144.219 ,127.375 
]

Kirundo_chirps_pre = [90.127 ,88.764 ,141.583 ,170.743 ,93.503 ,12.713 ,3.358 ,21.957 ,58.496 ,114.576 ,141.455,103.487 ]

Muyinga_chirps_prec = [110.873 ,99.919 ,146.244,171.013 ,84.084 ,6.595 ,1.700,14.823,52.508,109.087,140.261,123.473 
 ]

Kirundo_ERA5_pre = [89.469 ,87.688 ,129.938,174.125,94.188,13.781,3.688,19.125,61.813,128.375,156.750,114.844]

Muyinga_ERA5_prec = [132.219,122.813,178.375,204.594 ,99.844,12.125,3.156,17.500,64.594,148.750,193.313,162.125 
 
 ]
NEB_NDVI_mean = [0.592028464705882,0.548957405555556,0.565718622222222,0.623394438888889,0.644336927777778,0.58870065,0.496820788888889,0.435765227777778,0.424811105555556,0.469063211111111,0.582755555555556,0.637980294444444
]

#assign lists to a value
data = {'Month': Month, 'Kirundo_obs_pre': Kirundo_obs_pre, 'Muyinga_obs_prec': Muyinga_obs_prec, 'Kirundo_chirps_pre':Kirundo_chirps_pre,
       'Muyinga_chirps_prec': Muyinga_chirps_prec,'Kirundo_ERA5_pre':Kirundo_ERA5_pre,'Muyinga_ERA5_prec':Muyinga_ERA5_prec , 'NEB_NDVI_mean':NEB_NDVI_mean}
#convert dictionary to a dataframe
df = pd.DataFrame(data)
#Print out all rows
df[:12]
```

```python
# create line plot for average percipitation levels (Kirundo)
#plt.title('Average Percipitation Percentage by Month')
#sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.lineplot(x='Month', y='Kirundo_obs_pre', data=df, sort=False)
sns.lineplot(x='Month', y='Kirundo_chirps_pre', data=df, sort=False)
sns.lineplot(x='Month', y='Kirundo_ERA5_pre', data=df, sort=False)
plt.ylabel('Rainfall (mm)',fontsize=16)
plt.yticks(rotation=90)

plt.legend(['OBS', 'CHIRPS','ERA5'], loc=3)
plt.twinx()
sns.lineplot(x='Month', y='NEB_NDVI_mean', data=df, sort=False, color = 'black')
plt.yticks(rotation=90)
plt.xlabel(' ')
plt.ylabel('NDVI mean',fontsize=16)
plt.legend(['NDVI'], loc=4)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.savefig('/home/pbatungwanay/Bureau/precip_cycle_kirundo.eps')
```
