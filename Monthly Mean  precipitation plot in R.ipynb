```r
Kirundo = as.data.frame(Data$OBS_KIR)

                                         
 monthly_Kirundo=NULL
                                          
 for (i in 1:384)
                                          
   {
                                            
   k=Kirundo[i,]
                                            
   monthly_Kirundo =c(monthly_Kirundo,as.numeric(k) )
                                          }
#monthly_Kirundo
Monthly_kir=ts(monthly_Kirundo, start=c(1986,1), frequency=12)
Monthly_kir

#Monthly_kir.to_csv(r'/home/pbatungwanay/Bureau/Monthly_kir.csv', index =True ) # saving the monthly data
write.csv(Monthly_kir,"/home/pbatungwanay/Bureau/Monthly_kir_OBS.csv", row.names = FALSE)
```

# Regression plot

For Kirundo

```r
Data = read.csv('MONTHLY_CHIRP_KIR_MUY_AGREG2.csv')
```

```r
fit=lm(CHIRP_KIR~OBS_KIR,data=Data)
fit
```

```r
fit2=lm(ERA5_KIR~OBS_KIR,data=Data)
fit2
```

```r
summary(fit)
```

```r
summary(fit2)
```

```r


plot(Data$ERA5_KIR, Data$OBS_KIR, pch = 16, col = 'red',cex = 1.8, xlab = '', lty=3,las=0,lwd = 2 ,
     ylab = 'Monthly rainfall (mm)', cex.lab = 1.5, cex.axis = 1.2)
par(new=T)
plot(Data$CHIRP_KIR, Data$OBS_KIR, pch = 19, cex = 1.8, axes=F, ylab = '',xlab = '')
text(x=120, y= 350, labels = expression('y'*'='*'mx' *'+'* 'c'),cex = 1.2)
text(x=120, y= 330, labels = expression('R'*'='*'0.77'~ (ERA5 )),cex = 1.2)
text(x=125, y= 310, labels = expression('R'*'='*'0.84'~ (CHIRPS )),cex = 1.2)

abline(fit2, col = 'red',cex = 2, lwd = 5)
abline(fit, col = 'black',cex = 2, lwd = 5)
legend("topleft", legend=c("Era5","Chirps"), pch=c(16,19),col=c("red","black"),cex = 1.2)
dev.copy(jpeg,'regression_kir.jpeg')
dev.off()
```

For Muyinga

```r
fit_muy=lm(CHIRP_MUY~OBS_MUY,data=Data)
fit_muy2=lm(ERA5_MUY~OBS_MUY,data=Data)
plot(Data$ERA5_MUY, Data$OBS_MUY, pch = 16, col = 'red',cex = 1.8, xlab = '', 
     ylab = 'Monthly rainfall (mm)', , cex.lab = 1.5, cex.axis = 1.2)
par(new=T)
plot(Data$CHIRP_MUY, Data$OBS_MUY, pch = 16, cex = 1.8, axes=F, ylab = '', xlab = '')
text(x=40, y= 290, labels = expression('y'*'='*'mx' *'+'* 'c'),cex = 1.2)
text(x=40, y= 270, labels = expression('R'*'='*'0.81'~ (ERA5 )),cex = 1.2)
text(x=45, y= 250, labels = expression('R'*'='*'0.88' ~(CHIRPS )),cex = 1.2)

abline(fit_muy2, col = 'red',cex = 2,lwd = 5)
abline(fit_muy, col = 'black',cex = 2,lwd = 5)
legend("topleft", legend=c("Era5","Chirps"), pch=c(16,19),col=c("red","black"),cex = 1.2)
dev.copy(png,'regression_muy.png')
dev.off()
```
